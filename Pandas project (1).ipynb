```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

This line is **cleaning and formatting column names** in a DataFrame (`df`) ‚Äî a super important step in data cleaning, especially for professional projects.

---

### üîç What does it do?

You're chaining string operations to **standardize column names** in a format called **`snake_case`**, which is clean and Python-friendly.

Let‚Äôs break it into parts:

---

### üß± 1. `df.columns`
This accesses the **column names** of your DataFrame.

Example before cleaning:

```python
[' Show ID ', 'Release Year', 'Views']
```

---

### üßπ 2. `.str.strip()`
This **removes leading and trailing spaces** from each column name.

```python
['Show ID', 'Release Year', 'Views']
```

---

### üî° 3. `.str.lower()`
This converts **all letters to lowercase** ‚Äî helps avoid case sensitivity bugs.

```python
['show id', 'release year', 'views']
```

---

### üêç 4. `.str.replace(' ', '_')`
This replaces all **spaces** with **underscores**, making the names **snake_case** (preferred in Python).

```python
['show_id', 'release_year', 'views']
```

---

### ‚úÖ Final Result:
```python
['show_id', 'release_year', 'views']
```

Now these column names are:
- Clean
- Consistent
- Easy to access in code (e.g., `df.show_id` or `df['show_id']`)

---

### üß† Why is this useful?

‚úÖ Prevents bugs  
‚úÖ Makes your code readable and consistent  
‚úÖ Follows Python best practices (snake_case)

---

### üéØ In short:
This line of code is a **one-liner magic wand** that:
- Trims spaces
- Lowers case
- Replaces spaces with underscores

Perfect for professional data analysis projects! üöÄ

2) Average Duration of Shows per Country

3. Genre with the Highest Total Views
To find the genre with the highest total views:

````python
genre_views = df.groupby('GENRE')['VIEWS'].sum()
top_genre = genre_views.idxmax()
print(top_genre)
````

````python
# 4.Filter Shows with Duration > 50 Minutes and Views > 1 Million.
filtered_shows = df[(df['duration_minutes'] > 50) & (df['VIEWS'] > 1_000_000)]
print(filtered_shows[['TITLE','DURATION','VIEWS']])
````

````python
# 5.Countries with More Than One Show
country_counts  = df['COUNTRY'].value_counts()
countries_with_multiple_shows = country_counts[country_counts > 1]
print(countries_with_multiple_shows)
````

````python
# 7 Find Shows with Titles Containing the Word "in"
shows_with_in = df[df['TITLE'].str.contains(r'\bdark\b', case=False, regex=True)]
print(shows_with_in[['TITLE']])
````

This line filters the DataFrame df to include only those rows where the title column contains the word "in" as a standalone word, not as part of another word. Here's a step-by-step explanation:

df['title']: Accesses the 'title' column of the DataFrame, which is a Series containing the titles of the shows.‚Äã

.str.contains(...): Applies a string method to each element in the Series to check if it contains a specified pattern.‚Äã

r'\bin\b': This is a regular expression pattern. Let's dissect it:

r'': The r prefix denotes a raw string in Python, which tells Python not to interpret backslashes as escape characters. This is important for regular expressions where backslashes are common.

\b: In regular expressions, \b represents a word boundary. It matches the position between a word character (like letters and numbers) and a non-word character (like spaces or punctuation).

in: The literal string we're searching for.

\b: Another word boundary to ensure that "in" is matched as a whole word.‚Äã
